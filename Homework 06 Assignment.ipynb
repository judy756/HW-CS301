#### Question 2 - Create a new dataframe which is a copy of the orginal one. In this copy, replace the missing values in the data set using one of the discussed methods in class.

Replacing missing value with the mean

```python
copy = df.copy()
for col in copy.columns:
    empty = copy.loc[copy[col].isna()];
    if empty.shape[0]!=0:
        copy.loc[df[col].isna(), col] = copy[col].mean()
```

#### Question 4 - Using the original dataframe (unchanged from Questoins 2, 3), create a mapping for the "Punxsutawney Phil" column which changes the categories into numerical quantities. #### Question 6 - For the remaining rows which have no record (but the other columns *are* populated), determine an appropriate method for inputting the value for this column. Explain your reasoning, and show your work below (using pandas).

```python
norecord = df.loc[df['Shadow'] == 0]
nosha = df.loc[df['Shadow'] == 1]
part = df.loc[df['Shadow'] == 2]
full = df.loc[df['Shadow'] == 3]
col1 = 'February Average Temperature'
col2 = 'March Average Temperature'
plt.plot(norecord[col1], norecord[col2], 'o', c='blue', label='No Record')
plt.plot(nosha[col1], nosha[col2], 'x', c='red', label='No Shadow')
plt.plot(part[col1], part[col2], 'o', c='black', label='Partial Shadow')
plt.plot(full[col1], full[col2], 'x', c='green', label = 'Full Shadow')
plt.gcf().set_size_inches(12, 6)
plt.xlabel('February')
plt.ylabel('March')
plt.title('Average Temperature of February and March')
plt.legend()
plt.show()
```

```python
norecord1 = df.loc[(df['Shadow'] == 0) | (df['Shadow'] == 1)]
norecord2 = df.loc[(df['Shadow'] == 0) | (df['Shadow'] == 2)]
norecord3 = df.loc[(df['Shadow'] == 0) | (df['Shadow'] == 3)]
col1 = 'February Average Temperature'
col2 = 'March Average Temperature'
plt.plot(norecord1[col1].mean(), norecord1[col2].mean(), 'o', c='blue', label='No Record and NoShadow')
plt.plot(nosha[col1].mean(), nosha[col2].mean(), 'x', c='blue', label='Only No Shadow')
plt.plot(norecord2[col1].mean(), norecord2[col2].mean(), 'o', c='red', label='No Record and PartShadow')
plt.plot(part[col1].mean(), part[col2].mean(), 'x', c='red', label='Only Partial Shadow')
plt.plot(norecord3[col1].mean(), norecord3[col2].mean(), 'o', c='green', label='No Record and FullShadow')
plt.plot(full[col1].mean(), full[col2].mean(), 'x', c='green', label = 'Only Full Shadow')
plt.gcf().set_size_inches(12, 6)
plt.xlabel('February')
plt.ylabel('March')
plt.title('Mean Average Temperature of February and March')
plt.legend()
plt.show()
```

As shown in the two graphs above, no record is best to be incorporate into full shadow.

The size of Partial Shadow is too small to see a pattern. No shadow seem to have a greater value in March and February than the no record. Full shadow is more scatter all over the graph.
